### Mapping Diagnosis Values to Binary
Benign as 0 and Malignant as 1.

```python
dataset['diagnosis'] = dataset['diagnosis'].map({'B':0,'M':1}).astype(int)
dataset.head()
```

### Segregate Data into INPUT X and OUTPUT Y

```python
X = dataset.iloc[:,2:32].values
X
```

```python
Y = dataset.iloc[:,1].values
Y
```

### Splitting Data into Train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Validating some ML algorithm by its accuracy-Model Score

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
models = []

models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', LinearDiscriminantAnalysis()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))

print(models)
```

```python
results = []
names = []
res = []
for name, model in models:
  kfold = StratifiedKFold(n_splits=10, random_state= None)
  cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
  results.append(cv_results)
  names.append(name)
  res.append(cv_results.mean())
  print('%s: %f' % (name, cv_results.mean()))


pyplot.ylim(.900, .999)
pyplot.bar(names, res, color='maroon', width =0.6)

pyplot.title('Algorithm Comparison')
pyplot.show()
```

Highest Accuracy Model is the Logistic Regression Model.

### Training and Prediction using highest accuracy model
